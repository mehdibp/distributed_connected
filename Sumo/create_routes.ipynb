```python
import subprocess
import os

# os.environ["SUMO_HOME"] = "D:/Applications/Sumo"                                        # Set the SUMO_HOME path
random_trips_path = os.path.join(os.environ["SUMO_HOME"], "tools", "randomTrips.py")    # Path to randomTrips.py file

# Execute command
subprocess.run([
    "python", random_trips_path,
    "-n", "./Networks/Zanjan_1.net.xml",
    "-o", "./Networks/Zanjan_1.trips.xml",
    "--seed", "42",
    "--prefix", "veh",
    "--trip-attributes", 'departLane="best" departSpeed="max" departPos="random"',
    "--end", "2000",
    "--period", "1",
    "--binomial", "100",
    "--fringe-factor", "2",
    "--validate"
])

# Run duarouter to create the rou.xml file
subprocess.run([
    "duarouter",
    "-n", "./Networks/Zanjan_1.net.xml",
    "-t", "./Networks/Zanjan_1.trips.xml",
    "-o", "./routes/Zanjan_1.rou.xml",
    "--ignore-errors"
])
```
